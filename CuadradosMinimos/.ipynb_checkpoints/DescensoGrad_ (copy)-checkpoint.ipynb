```python
import random
import matplotlib.pyplot as plt
import numpy as np
 
def gen_data(n, bias, varianza):
    x = []
    y = []
    z = []
    for i in range(0, n):
        x.append(i)
        y.append((i + bias) + random.uniform(0, 1) * varianza)
    return x, y
 
x,y = gen_data(100, 25, 50)
plt.scatter(x, y)
```

```python
def coste(x,y,a,b):
  m=len(x)
  error=0.0
  for i in range(m):
    hipotesis=a+b*x[i] ## y_g
    error+=(y[i]-hipotesis)**2

  return error/(2*m)
```

```python
def descenso_gradiente(x,y,a,b,alpha,epochs):
  # a es ordenada al origen
  # b es la pendiente
  # alpha es la taza de aprendizaje
  m=len(x) # cantidad de muestras

  hist_coste=[]
  for ep in range(epochs):
    b_deriv=0.0; a_deriv=0.0;
    for i in range(m):
      hipotesis=a+b*x[i]
      a_deriv += hipotesis-y[i]
      b_deriv +=  (hipotesis-y[i])*x[i]
      hist_coste.append(coste(x,y,a,b))

    a -= (a_deriv/m)*alpha
    b -= (b_deriv/m)*alpha

  return a,b,hist_coste
```

```python
a=1; b=1; alpha=0.0001; epochs= 100000;
a,b,historial=descenso_gradiente(x,y,a,b,alpha,epochs)
```

```python
import numpy as np 

ite=np.arange(1,len(historial)+1)
plt.plot(ite,historial)
#plt.ylim([0,150])
plt.show()
```

```python
y_new=b*np.array(x)+a

plt.scatter(x, y,color='red')
plt.plot(x,y_new,color='black')
plt.grid()

a,b
```
